```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [16,9]
```

```python
Data = pd.read_csv('/Users/colinreid/Documents/GitHub/Explore-Chicago/CTA_-_Ridership_-_Bus_Routes_-_Daily_Totals_by_Route.csv',',')
Data['date_time'] = pd.to_datetime(Data['date'])
```

```python
Route = '79'
DayType = 'W'
Bus_Data = Data[Data['route'] == Route]
```

```python
fy11_all = Bus_Data[(Bus_Data['date_time'] >= '2011-01-01') & (Bus_Data['daytype'] == DayType) & (Bus_Data['date_time'] < '2012-01-01')][['date_time','rides']]
fy12_all = Bus_Data[(Bus_Data['date_time'] >= '2012-01-01') & (Bus_Data['daytype'] == DayType) & (Bus_Data['date_time'] < '2013-01-01')][['date_time','rides']]
fy13_all = Bus_Data[(Bus_Data['date_time'] >= '2013-01-01') & (Bus_Data['daytype'] == DayType) & (Bus_Data['date_time'] < '2014-01-01')][['date_time','rides']]
fy14_all = Bus_Data[(Bus_Data['date_time'] >= '2014-01-01') & (Bus_Data['daytype'] == DayType) & (Bus_Data['date_time'] < '2015-01-01')][['date_time','rides']]
fy15_all = Bus_Data[(Bus_Data['date_time'] >= '2015-01-01') & (Bus_Data['daytype'] == DayType) & (Bus_Data['date_time'] < '2016-01-01')][['date_time','rides']]
fy16_all = Bus_Data[(Bus_Data['date_time'] >= '2016-01-01') & (Bus_Data['daytype'] == DayType) & (Bus_Data['date_time'] < '2017-01-01')][['date_time','rides']]
fy17_all = Bus_Data[(Bus_Data['date_time'] >= '2017-01-01') & (Bus_Data['daytype'] == DayType) & (Bus_Data['date_time'] < '2018-01-01')][['date_time','rides']]
```

```python
ax = fy11_all.plot(x='date_time',y='rides',title=Route + ' Ridership',legend=False)
fy12_all.plot(x='date_time',y='rides',ax=ax,legend=False)
fy13_all.plot(x='date_time',y='rides',ax=ax,legend=False)
fy14_all.plot(x='date_time',y='rides',ax=ax,legend=False)
fy15_all.plot(x='date_time',y='rides',ax=ax,legend=False)
fy16_all.plot(x='date_time',y='rides',ax=ax,legend=False)
fy17_all.plot(x='date_time',y='rides',ax=ax,legend=False)
```
